Escrevendo na forma matricial temos:

\begin{equation}
    \left[\begin{array}{ccc} m & 0 & 0 \\ 0 & m & mx_g \\ 0 & mx_g & I_z  \end{array}\right] \left\{\begin{array}{c} \dot{u} \\ \dot{v} \\ \dot{r}  \end{array}\right\} + \left[\begin{array}{ccc} 0 & 0 & -m(x_gr+v) \\ 0 & 0 & mu \\  m(x_gr+v) & -mu & 0  \end{array}\right] \left\{\begin{array}{c} u \\ v \\ r  \end{array}\right\} =  \left\{\begin{array}{c} X \\ Y \\ N  \end{array}\right\}
\end{equation}

Podemos reescrever como:

\begin{equation}
    M \dot{\nu} + C(\nu)\nu = \tau,
\end{equation}

onde $M$ é a matriz de massa, $C$ é a matriz de Coriolis, $\nu=[u,v,r]^T$ é o vetor de velocidades e $\tau =  [X,Y,N]^T$ é o vetor de forças externas, que pode ser dividido em:

\begin{equation}
    \tau=\tau_{hyd}+\tau_{hs}+\tau_{wind}+\tau_{wave}+\tau_{control},
\end{equation}

onde $\tau_{hyd}$ são as forças hidrodinâmicas, $\tau_{hs}$ as forças hidroestáticas, $\tau_{wind}$ a força do vento, $\tau_{wave}$ a força das ondas e $\tau_{control}$ a força gerada pelos atuadores. 



É possível relacionar o sistema de coordenadas fixo no corpo com um sistema global conhecido como NED (\textit{north-east-down}). Esta relação para o movimento de três graus de liberdade de um navio é dada por:

\begin{equation}
    \dot{\eta}=R(\psi)\nu,
\end{equation}

onde $\eta=[N,E,\psi]^T$ é o vetor de posição  e  $R(\psi)=\left[\begin{array}{ccc}\cos \psi & -\sin \psi & 0 \\ \sin \psi & \cos \psi & 0 \\0 & 0 & 1\end{array}\right]$ é a matriz de rotação.

Seja $x$ um vetor de estados dado por 

\begin{equation}
    x = [u,v,r,N,E,\psi]^T
\end{equation}

A dinâmica de um navio considerando velocidade e posição pode ser dada por:

\begin{equation}
\left[\begin{array}{cc} M & 0_{3x3} \\ 0_{3x3} & I_{3x3} \end{array}\right]    \left\{\begin{array}{c} \dot{u} \\ \dot{v} \\ \dot{r} \\ \dot{N} \\ \dot{E} \\ \dot{\psi}  \end{array}\right\} = \left[\begin{array}{cc} -C(\nu) & 0_{3x3} \\ R(\psi) & 0_{3x3} \end{array}\right]  \left\{\begin{array}{c} u \\ v \\ r \\ N \\ E \\ \psi  \end{array}\right\} + \left[\begin{array}{c} I_{3x3} \\ 0_{3x3} \end{array}\right]   \tau 
\end{equation}

Reescrevendo:

\begin{equation}
\left\{\begin{array}{c} \dot{u} \\ \dot{v} \\ \dot{r} \\ \dot{N} \\ \dot{E} \\ \dot{\psi}  \end{array}\right\} = \left[\begin{array}{cc} -M^{-1}C(\nu) & 0_{3x3} \\ R(\psi) & 0_{3x3} \end{array}\right]  \left\{\begin{array}{c} u \\ v \\ r \\ N \\ E \\ \psi  \end{array}\right\} + \left[\begin{array}{c} M^{-1} \\ 0_{3x3} \end{array}\right]   \tau 
\end{equation}

Considerando como esforço de controle $u = \tau_{control}$ e as outras componentes de $\tau$ como distúrbio e considerando também que podemos medir os 6 estados, temos o sistema em espaço de estados na forma


$\begin{cases}
    \dot{x} = A(x) x + B(x) u + w_p \\
    y = I_{6x6} x + w_s \\
\end{cases}$

onde a matriz de estados $A(x)$ é dada por

\begin{equation}
   A(x) = \left[\begin{array}{cc} -M^{-1}C(\nu) & 0_{3x3} \\ R(\psi) & 0_{3x3} \end{array}\right]
\end{equation}

a matriz de entradas $B(x)$ é dada por

\begin{equation}
   B(x) = \left[\begin{array}{c} M^{-1} \\ 0_{3x3} \end{array}\right]
\end{equation}

o vetor de entradas $u$ é dado por

\begin{equation}
    \left[\begin{array}{c} X_{control} \\ Y_{control} \\ N_{control} \end{array}\right]
\end{equation}

e $w_p$ e $w_s$ são, respectivamente, os distúrbios da planta e dos sensores.
